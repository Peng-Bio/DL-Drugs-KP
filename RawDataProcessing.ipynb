```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/mediAI/DDI-CNN/data/drug_information.csv')
df.head()
```

```python
print(f'Number of unique drug: {len(df)}')
drug_name = df.drug_name.unique()
print(f'Number of unique drug: {len(drug_name)}')
```

```python
df_event = pd.read_csv('/content/drive/MyDrive/mediAI/DDI-CNN/data/ddi_biotech 022423 Format.csv')
df_event.head()
```

```python
# Filter out the drugs which we are not interested in.
print(f'Number of df_event:{df_event.shape}')
df_event_new = df_event[df_event['drug1'].isin(drug_name) & df_event['drug2'].isin(drug_name)]
print(f'Number of df_event_new:{df_event_new.shape}')

df_event_new.to_csv('/content/drive/MyDrive/mediAI/DDI-CNN/data/event_new.csv',index=False)
```

```python
df_data = pd.read_csv('/content/drive/MyDrive/mediAI/DDI-CNN/data/drugbank_data.csv')
df_data.head()
```

```python
print(f'Number of df_event:{df_data.shape}')
df_data_new = df_data[df_data['Name'].isin(drug_name)]
print(f'Number of df_event_new:{df_data_new.shape}')
```

```python
df_data.rename(columns={"Name": "drug_name"},inplace=True)
df_data.head()
```

```python
df_data_new = pd.merge(df, df_data,how="left",on=['drug_name'])
df_data_new.drop(columns=['index','drugbank_id','SMILES','SNP Adverse Drug Reactions'],inplace=True)
df_data_new.head()
```

```python
# Replace the empty cell to "unknown"
df_data_new = df_data_new.replace('[]', '[unknown]')
df_data_new.head()
```

```python
df_data_new.isnull().sum()
```

```python
df_data_new.fillna(value='unknown',inplace=True)
df_data_new.tail()
```

```python
df_data_new.to_csv('/content/drive/MyDrive/mediAI/DDI-CNN/data/data_new.csv',index=False)
```
